# RESULTS

```python
import torch
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt

# Load the custom-trained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp/weights/best.pt', force_reload=True)

# Function to detect fabric defects
def detect_fabric_defects(image_path):
    # Load the image
    img = cv2.imread(image_path)
    
    # Make detections
    results = model(img)
    
    # Extract predictions (boxes, labels, and confidence scores)
    pred = results.pred[0]
    
    # If no objects are detected, mark it as defect-free
    if pred is None or len(pred) == 0:
        print(f"{image_path} - Fabric is defect free.")
        cv2.putText(img, 'defect free', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)  # Green text
    else:
        print(f"{image_path} - Fabric has defects.")
        cv2.putText(img, 'defect', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)  # Red text
    
    # Display the results with defect class labels (if any)
    for *xyxy, conf, cls in pred:
        label = results.names[int(cls)]  # Use the class names from the YAML file
        confidence = round(conf.item(), 2)
        class_name = label.lower()  # Assuming classes are 'hole', 'stain', etc.
        
        # Add label with confidence score
        cv2.putText(img, f'{class_name} {confidence}', (int(xyxy[0]), int(xyxy[1])), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)  # White text for label
        cv2.rectangle(img, (int(xyxy[0]), int(xyxy[1])), (int(xyxy[2]), int(xyxy[3])), (0, 255, 0), 2)  # Green bounding box
    
    # Display the image
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Hide axis
    plt.show()

# Function to loop through images in the directory
def process_test_images(image_dir):
    # List all images in the test folder
    image_files = [f for f in os.listdir(image_dir) if f.endswith('.jpg') or f.endswith('.png')]  # Add file types as needed
    
    for image_file in image_files:
        image_path = os.path.join(image_dir, image_file)
        print(f"Processing {image_file}...")
        detect_fabric_defects(image_path)

# Path to your images folder
image_dir = r'C:\Users\shaiiikh\py\attempt\data\images\test'

# Process all images in the test folder
process_test_images(image_dir)
```
